# Data cleaning

## Preparing the Data for insertion into the database

First we need to import pandas and the csv file we are going to clean

```python
import pandas as pd
```

## Summary csv

We are starting with Quaterly greenhouse emissions csv in order to have a clean file to then create our relational database.

```python
df = pd.read_csv("1_Quarterly_Greenhouse_Gas_GHG_Air_Emissions_Accounts.csv")
df
```

First we are going to drop some rows to make it more workabe. We are only interested in carbon dioxide emissions and will only take into accound those who have been seasonally adjusted.

```python
df = df[df['Gas Type'] == 'Carbon dioxide']
df = df[df['Seasonal Adjustment'] == 'Seasonally Adjusted']
df
```

Then we are goind to drop the columns we will not need

```python
columns_to_drop = ["ISO2","ISO3","CTS Code","CTS Name", 'CTS Full Descriptor', 'Gas Type', 'Scale', 'Seasonal Adjustment', 'Unit', 'Source', 'Indicator']
df = df.drop(columns = columns_to_drop)
df
```

We only kept the country columns and the industry as well as the quarters columns as those are the ones that are going to be relevant for our analysis. We also are going to change the data from quarters to years because we want to analyse yearly emissions instead of quaterly emissions.

```python
# List of years to process
years = ['2010', '2011','2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022', ]

# Initialize an empty dictionary to hold the yearly sums
yearly_data = {}

# Sum the values of each year
for year in years:
    yearly_data[year] = df[[f'{year}Q1', f'{year}Q2', f'{year}Q3', f'{year}Q4']].sum(axis=1)

# Create a new DataFrame with the yearly summed values
df_yearly = pd.DataFrame(yearly_data)
df_yearly
```

```python

columns_to_drop = [col for col in df.columns if any(map(str.isdigit, col))]

df_country_industry = df.drop(columns=columns_to_drop)
df_country_industry
```

```python

df1 = df_country_industry
df2 = df_yearly
df = pd.concat ([df1, df2], axis=1)
df
```

```python
df.to_csv('Yearly_emissions.csv', index=False)
```
