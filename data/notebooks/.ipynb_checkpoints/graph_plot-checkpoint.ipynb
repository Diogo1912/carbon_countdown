# Plotting

## After looking through the database, the graphs that are interesting to plot in the context of the website have been chosen:

* Global emissions by industry over the years (line plot)
* Emissions across different regions and differences between 2010 and 2022 (bar chart)
* Comparison between advanced economies and emerging economies (line plot)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Yearly_emissions.csv')
df
```

```python
data = df
world_emissions = data[
    (data['Country'] == 'World') & 
    (data['Industry'] == 'Total Industry and Households')]
emissions_values = world_emissions.loc[:, '2010':'2022'].values.flatten()
emissions_values
```

```python
data = {
    'Year': [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022],
    'Emissions': [
        33671.29684, 34747.338642, 35314.593067, 35980.447011, 36323.18346, 
        36107.095479, 35496.221717, 36149.571035, 37065.885329, 37050.517853, 
        35224.444967, 37257.656362, 37590.882753
    ]
}
df = pd.DataFrame(data)


sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Emissions', data=df, marker='o')

plt.title('World Emissions for Total Industry and Households (2010-2022)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Emissions', fontsize=14)

plt.show()
```

```python
# Define the list of continent names
continents = ['Africa', 'Asia', 'Europe', 'Northern America', 'Latin America and the Caribbean', 'Oceania']

# Filter the data to include only the continents
data_continents = data[data['Country'].isin(continents)]

# Select only the relevant columns (2010 and 2022 emissions)
plot_data = data_continents[['Country', '2010', '2022']]
plot_data = plot_data.melt(id_vars=['Country'], var_name='Year', value_name='Emissions')

# Set up the plot
plt.figure(figsize=(12, 8))
sns.barplot(data=plot_data, x='Country', y='Emissions', hue='Year', edgecolor='none', errorbar=None)
plt.title('Emissions in 2010 vs 2022 by Continent')
plt.xlabel('Continent')
plt.ylabel('Emissions')
plt.xticks(rotation=45)  # Rotate continent names for better readability
plt.legend(title='Year')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Filter data to only include the year 2022 emissions, excluding 'Total Industry and Households'
industry_emissions_2022 = data[['Industry', '2022']]
industry_emissions_2022 = industry_emissions_2022[industry_emissions_2022['Industry'] != 'Total Industry and Households']  # Exclude the category

# Set up the plot
plt.figure(figsize=(12, 8))
sns.barplot(data=industry_emissions_2022, x='Industry', y='2022', palette='viridis', errorbar=None)

# Customize the plot
plt.title('Emissions by Industry in 2022 (Excluding Total Industry and Households)')
plt.xlabel('Industry')
plt.ylabel('Emissions')
plt.xticks(rotation=90)  # Rotate the industry names for better readability
plt.tight_layout()

# Show the plot
plt.show()
```
